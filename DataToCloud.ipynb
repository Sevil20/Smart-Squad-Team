```python
import time
import requests
import math
import random
import serial
TOKEN = "BBFF-W9BSo0JRtwjKdY2djoRKMG0hrHL8XF"  
DEVICE_LABEL = "My_PC"
VARIABLE_LABEL_1 = "temperature"
VARIABLE_LABEL_2 = "humidity"



ser = serial.Serial('COM3', 9600, timeout=1)
print("Reading data from serial port.....");
time.sleep(2)
ser.reset_input_buffer()

def build_payload(variable_1, variable_2, data):
   
    value_1 = float(data[0])
    value_2 = float(data[1])

    payload = {variable_1: value_1,variable_2: value_2}

    return payload


def post_request(payload):
    url = "http://industrial.api.ubidots.com"
    url = "{}/api/v1.6/devices/{}".format(url, DEVICE_LABEL)
    headers = {"X-Auth-Token": TOKEN, "Content-Type": "application/json"}

    status = 400
    attempts = 0
    while status >= 400 and attempts <= 5:
        req = requests.post(url=url, headers=headers, json=payload)
        status = req.status_code
        attempts += 1
        time.sleep(1)


    if status >= 400:
        print("[ERROR] Could not send data after 5 attempts, please check \
            your token credentials and internet connection")
        return False

    print("Device ID updated...wait for data transfer")
    return True


def main():
    line = ser.readline() 
    if line:
        rxdata = line.decode()
        tmp = rxdata.split(',')
        payload = build_payload(VARIABLE_LABEL_1, VARIABLE_LABEL_2,tmp)
        print(tmp[0],tmp[1])
        del tmp

        print("Data transfer started")
        post_request(payload)
        print("Data Successfully sent to cloud")


if __name__ == '__main__':
    while (True):
        main()
        time.sleep(1)
```
